### 清理数据

```python
clean_daf1=original_daf1.copy()
clean_daf1
```

### 把不合适数据进行更改id host_id longitude latitude 应为字符串 last_review 应为日期 reviews_per_month 应为整数 

```python
clean_daf1[["id","host_id","longitude","latitude"]]=clean_daf1[["id","host_id","longitude","latitude"]].astype(str)
```

```python
clean_daf1.info()
```

```python
clean_daf1["last_review"]=pd.to_datetime(clean_daf1["last_review"])
```

```python
clean_daf1.info()
```

### 把空值变为零再更改类型

```python
clean_daf1[clean_daf1["reviews_per_month"].isnull()]
```

```python
clean_daf1["reviews_per_month"]=clean_daf1["reviews_per_month"].replace(np.nan,0)
```

```python
clean_daf1["reviews_per_month"].info()
```

```python
clean_daf1["reviews_per_month"]=clean_daf1["reviews_per_month"].astype(int)
```

```python
clean_daf1.info()
```
